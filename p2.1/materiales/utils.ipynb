**numpy.argmax: Returns the indices of the maximum values along an axis. https://numpy.org/doc/stable/reference/generated/numpy.argmax.html**

```python
import numpy as np
array = np.random.rand(10)
print ("Sample array")
print (array)
idxmax = np.argmax(array)
print ("Index {} has the highest value ({}) of the array".format(idxmax, array[idxmax]))
```

**np.argsort: Returns the indices that would sort an array.**

```python
# Creando un array de muestra con 10 números aleatorios
array = np.random.rand(10)
print("Sample array:")
print(array)

# Utilizamos argsort para obtener los índices que ordenarían el array,
# y luego seleccionamos los últimos tres índices para obtener los tres valores más altos
idxs_ordenados = np.argsort(array)
top_3_idxs = idxs_ordenados[-3:]  # Índices de los tres valores más altos

# Imprimimos los índices y los valores correspondientes
print("Indices of the three highest values:", top_3_idxs)
print("Three highest values:", array[top_3_idxs])
```
